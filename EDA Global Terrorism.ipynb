## Exploratory Data Analysis on Dataset - Terrorism

```python
df.rename(columns = {'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'latitude',
    'longitude':'longitude','summary':'summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
     'nwound':'Wound'}, inplace = True)
```

```python
df = df[['Year','Month','day','Country','State','Region','City','latitude','longitude',"Attacktype",'kill',
               'Wound','target1','summary','Group','Targettype','Weapon','motive']]
```

```python
df.head(3)
```

```python
df.shape
```

```python
df['Wound'] = df['Wound'].fillna(0)
df['kill'] = df['kill'].fillna(0)
```

```python
df['Casualities'] = df['kill'] + df['Wound']
```

## Data Visualization

## Year Plot

```python
plt.subplots(figsize = (15,8))
sns.countplot('Year',data = df, palette = 'RdYlGn_r', edgecolor = sns.color_palette('dark',10))
plt.xticks(rotation = 45)
plt.title('No. Of Terrorist Activities Each Year')
plt.show()
```

```python
pd.crosstab(df.Year, df.Region).plot(kind = 'area', stacked = False, figsize = (15,8))
plt.title('Terrorist Activities By Region In Each Year',fontsize = 25)
plt.ylabel('Number of Attacks',fontsize = 25)
plt.xlabel("Year",fontsize = 25)
plt.show()
```

## Number of Total Attacks in Each Country

```python
country_attack = df.Country.value_counts()[:10]
country_attack
```

```python
df.Group.value_counts()[1:10]
```

```python
plt.subplots(figsize = (12,6))
country_attacks = df.Country.value_counts()[:10].reset_index()
country_attacks.columns = ['Country', 'Total Attacks']
sns.barplot(x = country_attacks.Country, y = country_attacks['Total Attacks'], palette = 'YlOrBr_r', edgecolor = sns.color_palette('dark',10))
plt.xticks(rotation = 45)
plt.title('Number of Total Attacks in Each Country')
plt.show()
```

## People died due to Attack

```python
df1 = df[['Year','kill']].groupby(['Year']).sum()

fig, ax = plt.subplots(figsize = (15, 8))
df1.plot(kind = 'bar', alpha = 0.6, ax = ax)
plt.xticks(rotation = 45)
plt.title("People died due to Attack",fontsize = 20)
plt.ylabel("Number of killed peope", fontsize = 20)
plt.xlabel('Year',fontsize = 20)

top_side = ax.spines["top"]
top_side.set_visible(False)
right_side = ax.spines["right"]
right_side.set_visible(False)

```

## Top 15 most affcted city

```python
df['City'].value_counts().to_frame().sort_values('City',axis = 0, ascending = False).head(15).plot(kind = 'bar',figsize = (15,8), color = 'Lightpink')
plt.xticks(rotation = 45)
plt.xlabel("City",fontsize = 20)
plt.ylabel("Number of attack",fontsize = 20)
plt.title("Top 15 most effected city",fontsize = 20)
plt.show()
```

```python
## Name of Attacktype

df['Attacktype'].value_counts().plot(kind = 'bar',figsize = (15,8), color = 'darkslateblue')
plt.xticks(rotation = 45)
plt.xlabel("Attacktype",fontsize = 15)
plt.ylabel("Number of attack",fontsize = 15)
plt.title("Name of attacktype",fontsize = 15)
plt.show()
```

## Number of Wound

```python
df[['Attacktype','Wound']].groupby(["Attacktype"], axis = 0).sum().plot(kind = 'bar', figsize = (15,8), color = ['cyan'])
plt.xticks(rotation = 45)
plt.title("Number of wounded", fontsize = 15)
plt.ylabel('Number of people',fontsize = 15)
plt.xlabel('Attack type',fontsize = 15)
plt.show()
```

```python
df[['Attacktype','kill']].groupby(["Attacktype"], axis = 0).sum().plot(kind = 'bar',figsize = (15,8),color = ['Pink'])
plt.xticks(rotation = 45)
plt.title("Number of killed", fontsize = 15)
plt.ylabel('Number of people',fontsize = 15)
plt.xlabel('Attack type',fontsize = 15)
plt.show()
```

```python
plt.subplots(figsize = (15,8))
sns.countplot(df["Targettype"], order = df['Targettype'].value_counts().index, palette = "gist_heat", edgecolor = sns.color_palette("mako"));
plt.xticks(rotation = 90)
plt.xlabel("Attacktype",fontsize = 15)
plt.ylabel("count",fontsize = 15)
plt.title("Attack per year",fontsize = 15)
plt.show()
```

```python
df['Group'].value_counts().to_frame().drop('Unknown').head(10).plot(kind = 'bar',color = 'Yellow', figsize = (15,8))
plt.title("Top 10 terrorist group attack", fontsize = 15)
plt.xlabel("terrorist group name", fontsize = 15)
plt.ylabel("Attack number",fontsize = 15)
plt.show()
```
